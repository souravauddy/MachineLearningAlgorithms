```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
plt.scatter(x=X, y=y)
weights = (4.200998702583868, 2.7825606729058734)
predictions = []
previous_y = y

for row in X:
    y = weights[0]

    for x in row:
        y += x * weights[1]

    predictions.append(y)

predictions = np.array(predictions)
predictions.reshape(X.shape)

plt.plot(X, predictions)
plt.show()
```
